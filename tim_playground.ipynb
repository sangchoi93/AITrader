```python
import numpy as np
import time
import csv
from datetime import datetime
from scipy.interpolate import UnivariateSpline

class stonks:
    def __init__(self, starting_monies, ticker):
        self.money = starting_monies
        self.start = starting_monies
        self.ticker = ticker
        self.stonks = []
        self.bought = 0
    
    def buy(self, price, amount):
        for i in range(0,amount):
            if price < self.money:
                print('buy {} at {}'.format(self.ticker, price))
                self.stonks.append(price)
                self.money -= price
                self.bought += 1
            else:
                print('Not Enough money to buy')
    
    def sell(self, price):
        if self.stonks:
            gain = price-self.stonks[0]
            self.money += price
            self.stonks.pop(0)
            print('sell {} at {} for {} gain'.format(self.ticker, price, gain))
        else:
            print('no stocks to sell')
            
    def sellall(self, price):
        if self.stonks:
            gain = price*len(self.stonks)-sum(self.stonks)
            self.money += price*len(self.stonks)
            self.stonks = []
            print('sell all {} at {} for {} gain'.format(self.ticker, price, gain))
        else:
            print('cant sell all, no stocks to sell')
    
    def portfolio_summary(self):
        print('Bought {} stocks of {}'.format(self.bought, self.ticker))
        num_sold = self.bought - len(self.stonks)
        print('Sold {} stocks of {}'.format(num_sold, self.ticker))
        holdings = sum(self.stonks)
        gain = self.money + holdings - self.start
        print('Holding {} stocks of {} for total ${}'.format(len(self.stonks), self.ticker, holdings))
        print('Total money ${}, total gain ${}'.format(self.money, gain))
        
        
TIC = 'SPY'
stocks = stonks(10000, TIC)
cp = float(rs.get_latest_price(TIC)[0])
data_sp = np.array([cp])
time_sp = np.array([int(time.time())])
x = list(range(1,15))
fn = TIC + '_' + str(time_sp[0]) + '.csv'
trend = 'flat'
with open(fn, 'w') as csvf:
    writer = csv.writer(csvf)
    while True:
        cur_price = float(rs.get_latest_price(TIC)[0])
        print(cur_price)
        tim = int(time.time())
        writer.writerow([tim, cur_price])
        data_sp = np.append(data_sp,cur_price)
        time_sp = np.append(time_sp, tim)
        if data_sp.shape[0]>15:
            print(datetime.now())
            y = data_sp[-14:]
            #x = time_sp[-14:]
            y_spl = UnivariateSpline(x,y, k=4, s=2)
            dx = y_spl.derivative()
            roots = dx.roots()
            if len(roots)>0:
                d2x = y_spl.derivative(n=2)(roots)
                if d2x[-1]>0:
                    if trend == 'up':
                        print('positive inflection, trigger buy at ' + str(cur_price))
                        stocks.buy(cur_price, 5)
                    else:
                        print('positive inflection, setting up trend')
                        trend = 'up'
                else:
                    if trend == 'down':                     
                        print('negative inflection, trigger sell at ' + str(cur_price))
                        stocks.sellall(cur_price)
                        stocks.portfolio_summary()
                    else:
                        print('negative inflection, setting down trend')
                        trend = 'down'
            else:
                #No roots detected, we are likely in an up/down trend
                if all(val<0 for val in dx(x[-3:])):
                    print('General downtrend detected, sell')
                    stocks.sellall(cur_price)
                elif all(val>0 for val in dx(x[-3:])):
                    print('General uptrend detected, buy single stock')
                    stocks.buy(cur_price, 1)
            grad1 = np.gradient(y)
            grad2 = np.gradient(grad1)            
            plt.plot(x, y, 'ro', ms=5)
            plt.plot(x,y_spl(x), 'g')
            plt.show()

        time.sleep(5)
```
